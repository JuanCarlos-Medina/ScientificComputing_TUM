# Direction Fields of ODE's



## Plotting functions

-plotDirectionField- Simple plotting of direction field. The "quiver" function displays the vectors on the plot.


-vectorField-  Defines the ODE $$y'(t) = \lambda*y(t)^2+\mu*y(t)-\nu$$

         Arguments:
         y : vector of variables
         p : vecor of parameters [lam, mu, nu]

```python
import matplotlib.pyplot as plt
%matplotlib inline  
import numpy as np

def plotDirectionField(X, Y, U, V, pname):
    M = sqrt(pow(U, 2) + pow(V, 2))
    U = np.divide(U,M)
    V = np.divide(V,M)
    Q = plt.quiver( X, Y, U, V, units='xy', pivot='mid', width=0.06, scale=1/.5)
    title(pname)

def vectorField(y, p):
    lam, mu, nu = p
    return lam*Y*Y + mu*Y - nu
```

### Saddle Point:  $\mu$ = 0, $\nu$ = 0

Critical point at 0


```python
X,Y = meshgrid( arange(0,20.1,1.),arange(-1.,1.1,.1) )
lam = 1.
mu = 0.
nu = 0.

U = np.copy(X)
U.fill(1)
V = vectorField(Y,[lam,mu,nu])

plotDirectionField(X, Y, U, V, "Saddle Point")
```

### Two critical points: $\lambda$ = 5, $\mu$ = 4, $\nu$ = 1

```python
X,Y = meshgrid( arange(0,20.1,1.),arange(-2.,1.1,.2) )
lam = 5.
mu = 4.
nu = 1.

U = np.copy(X)
U.fill(1)
V = vectorField(Y,[lam,mu,nu])
figure()
plotDirectionField(X, Y, U, V, "Two Critical Points")
```

### $\lambda$ = 0, $\mu$ = 1

We obtain $y(t) = \frac{\nu}{\mu}$. With positive $\mu$: Unstable point

```python

X,Y = meshgrid( arange(0,20.1,1.),arange(0,2.1,.1) )
lam = 0.
mu = 1.
nu = 1.

U = np.copy(X)
U.fill(1)
V = vectorField(Y,[lam,mu,nu])
figure()
plotDirectionField(X, Y, U, V, "Lambda = 0, mu = 1")
```

### $\lambda$ = 0, $\mu$ = -1

We obtain $y(t) = \frac{\nu}{\mu}$. With negative $\mu$: Stable point

```python

X,Y = meshgrid( arange(0,20.1,1.),arange(-2.0,-0.01,.1) )
lam = 0.
mu = -1.
nu = 1.

U = np.copy(X)
U.fill(1)
V = vectorField(Y,[lam,mu,nu])
figure()
plotDirectionField(X, Y, U, V, "Lambda = 0, mu = -1")
```

### $\lambda$= 0, $\mu$ = 0

if $\nu$ is different than zero, the equality $\frac{dy}{dt}$ = 0 can never be reached

```python
X,Y = meshgrid( arange(0,20.1,1.),arange(-1.0,1.1,.1) )
lam = 0.
mu = 0.
nu = 1.

U = np.copy(X)
U.fill(1)
V = vectorField(Y,[lam,mu,nu])
figure()
plotDirectionField(X, Y, U, V, "Lambda = 0, mu = 0")
```
